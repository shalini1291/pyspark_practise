```python
#to get avg rating
from pyspark.sql import SparkSession
spark=SparkSession\
      .builder\
      .appName('some eg')\
      .config('spark.some.config.option','some-value')\
      .getOrCreate()
eg=spark.sparkContext.parallelize(['JAN, NY, 3',
'JAN, PA, 1',
'JAN, NJ, 2',
'JAN, CT, 4',
'FEB, PA, 1',
'FEB, NJ, 1',
'FEB, NY, 2',
'FEB, VT, 1',
'MAR, NJ, 2',
'MAR, NY, 1',
'MAR, VT, 2',
'MAR, PA, 3'])
eg1 = eg.map(lambda x: (x.split(',')[0].strip(), (int(x.split(',')[2].strip()), 1)))

# Aggregate the ratings and count the occurrences for each month
eg2 = eg1.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))

# Compute the average rating for each month
eg3 = eg2.map(lambda x: (x[0], x[1][0] / x[1][1])).collect()

# Print the results
print(eg3)

```
